```python
from collections import deque

def solve_puzzle(start, end):
    queue = deque([(start, [])])
    
    while queue:
        current, path = queue.popleft()
        
        if current == end:
            return path
        
        blank_spot = current.index(0) 
        possible_moves = [] 
        if blank_spot - 3 >= 0: possible_moves.append(blank_spot - 3)
        if blank_spot + 3 < 9: possible_moves.append(blank_spot + 3)
        if blank_spot % 3 != 0: possible_moves.append(blank_spot - 1)
        if (blank_spot + 1) % 3 != 0: possible_moves.append(blank_spot + 1)
        
        for move in possible_moves:
            new_state = current[:] 
            new_state[blank_spot], new_state[move] = new_state[move], new_state[blank_spot] 
            if new_state not in queue:
                queue.append((new_state, path + [new_state])) 
                
    return None 

start_state = [1, 2, 3, 8, 0, 4, 7, 6, 5]
end_state = [2, 8, 1, 0, 4, 3, 7, 6, 5]

solution = solve_puzzle(start_state, end_state)

if solution:
    print("Solution found in", len(solution), "moves:")
    for i, state in enumerate(solution):
        
        print("Move", i, ": ", state)
else:
    print("No solution found.")
```

```python
jug1_capacity = 4 
jug2_capacity = 3
target = 2

def transfer_water(jug1, jug2):
  transfer_amount = min(jug1, jug2_capacity-jug2)
  jug1 -= transfer_amount
  jug2 += transfer_amount
  return jug1, jug2
 
def solve_problem(capacity1, capacity2, target):
  initial_state = (0,0) 
  states = [initial_state]
  
  while True:
    current_state = states.pop()

    if current_state[0] == target:
      return current_state

    new_state = (current_state[0], 0)  
    if new_state not in states:
      states.append(new_state)

    if current_state[0] < capacity1:
      new_state = (capacity1, current_state[1])  
      if new_state not in states:
        states.append(new_state)

    new_state = transfer_water(current_state[0], current_state[1])
    if new_state not in states:
      states.append(new_state)

    if current_state[0] > 0:
      new_state = (0, current_state[1])  
      if new_state not in states:
        states.append(new_state)

solution = solve_problem(jug1_capacity, jug2_capacity, target)  
print(solution)
```

```python
import sys

graph = [[0, 10 ,15 ,20], 
         [10, 0, 35, 25],
         [15 ,35, 0, 30],
         [20, 25, 30, 0]]

num_nodes = len(graph)

def tsp(start_node):
  
  visited = [False] * num_nodes
  visited[start_node] = True
  
  path = [start_node]
  
  curr_node = start_node
  curr_cost = 0
  
  while len(path) < num_nodes:
    
    next_node = -1
    min_cost = sys.maxsize
    
    for node in range(num_nodes):
      
      if visited[node]==False and graph[curr_node][node]>0:
          
        if graph[curr_node][node] < min_cost:
          min_cost = graph[curr_node][node]
          next_node = node
          
    visited[next_node] = True
    path.append(next_node)
    curr_cost += min_cost  
    curr_node = next_node
    
  return path, curr_cost

start = int(input("Enter start node: "))
path, cost = tsp(start) 
print("Path:", path)
print("Cost:", cost)
```
